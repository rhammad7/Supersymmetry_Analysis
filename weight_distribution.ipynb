# Weight Distribution for  N1 990 GeV ; gl 1000 GeV ; 100k events ; using CKKW

```python
import ROOT
import math

ROOT.gSystem.Load("libDelphes")

try:
    ROOT.gInterpreter.Declare('#include "classes/DelphesClasses.h"')
    ROOT.gInterpreter.Declare('#include "external/ExRootAnalysis/ExRootTreeReader.h"')
except:
    pass

# Set the path to the ROOT file
# /data/analysis/iduminic/Hammad/MSSM_LO_gogo_TESTCKKW_V2-990-1000/Events/run_100k

home_dir = "root_files/N990_G1000/"
inputFile = home_dir + "tag_1_delphes_events.root"

# Create chain of root trees
chain = ROOT.TChain("Delphes")
chain.Add(inputFile)

# Create object of class ExRootTreeReader
treeReader = ROOT.ExRootTreeReader(chain)
numberOfEntries = treeReader.GetEntries()

# Get pointer to the branch containing Particle information
branchMET = treeReader.UseBranch("MissingET")
branchEvent = treeReader.UseBranch("Weight")

canvas1 = ROOT.TCanvas("canvas1", "Histogram", 800, 600)
histogram1 = ROOT.TH1F("histogram1", "Weighted Distribution", 20, 0, 1)
histogram_met1 = ROOT.TH1F("histogram_met1", "n", 20, 0, 1600)
for entry in range(numberOfEntries):
    treeReader.ReadEntry(entry)
    met = branchMET.At(0).MET
    histogram_met1.Fill(met)
    if branchMET.At(0).MET > 800:
        weight = branchEvent.At(6).Weight
        histogram1.Fill(weight)

histogram1.Draw()
histogram_met1.Draw()
canvas1.Draw()
```

```python
import ROOT
import math

ROOT.gSystem.Load("libDelphes")

try:
    ROOT.gInterpreter.Declare('#include "classes/DelphesClasses.h"')
    ROOT.gInterpreter.Declare('#include "external/ExRootAnalysis/ExRootTreeReader.h"')
except:
    pass

# Set the path to the ROOT file ?????

home_dir = "root_files/N990/"
inputFile = home_dir + "tag_1_delphes_events.root"

# Create chain of root trees
chain = ROOT.TChain("Delphes")
chain.Add(inputFile)

# Create object of class ExRootTreeReader
treeReader = ROOT.ExRootTreeReader(chain)
numberOfEntries = treeReader.GetEntries()

# Get pointer to the branch containing Particle information
branchMET = treeReader.UseBranch("MissingET")
branchEvent = treeReader.UseBranch("Weight")


canvas = ROOT.TCanvas("canvas", "Histogram", 800, 600)
histogram = ROOT.TH1F("histogram", "Weighted Distribution", 20, 0, 1)
histogram_met = ROOT.TH1F("histogram_met", "n", 20, 0, 1600)
for entry in range(numberOfEntries):
    treeReader.ReadEntry(entry)
    met = branchMET.At(0).MET
    histogram_met.Fill(met)
    if branchMET.At(0).MET > 800:
        weight = branchEvent.At(6).Weight
        histogram.Fill(weight)

histogram.Draw()
histogram_met1.Scale(0.1)
histogram_met1.Draw()
histogram_met.Draw("SAME")
canvas.Draw()
canvas.SetLogy()

```
